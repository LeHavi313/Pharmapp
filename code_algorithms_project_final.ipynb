```python
import heapq

medicine = input("What medicine do you need? (Zinnat / Rivotril / Amoxicilina / Valium / Xanax / Morphine / Salbuthanol)")
barrio = input("What neighborhood are you in? (Neighborhood 1 / Neighborhood 2 / Neighborhood 3 / Neighborhood 4 / Neighborhood 5 / Neighborhood 6 / Neighborhood 7)")


class Edge():     
    def __init__(self, startVertex, targetVertex, weight):
        self.weight = weight
        self.startVertex = startVertex
        self.targetVertex = targetVertex

class Node():
    def __init__(self, name):
        self.name = name
        self.visited = False
        self.predecesor = None
        self.adjancencylist = []
        self.stocklist = []
        self.minDistance = float("inf")
    def __lt__(self, other):
        selfPriority = self.minDistance
        otherPriority = other.minDistance
        return selfPriority < otherPriority
    def calculate_shortest_path(self):
        queue = []
        with_medicine = []
        start = self
        start.minDistance = 0
        heapq.heappush(queue, start)
        while queue:
            actualVertex = heapq.heappop(queue)
            if medicine in actualVertex.stocklist:
                with_medicine.append(actualVertex)
            for edge in actualVertex.adjancencylist:
                u = edge.startVertex
                v = edge.targetVertex
                newDistance = u.minDistance + edge.weight
                if newDistance < v.minDistance:
                    v.predecesor = u
                    v.minDistance = newDistance
                    heapq.heappush(queue, v)
        return with_medicine
    def getShortestPathTo(self,targetVertex):
        path = []
        node=targetVertex
        while node is not None:
                path.append(node.name)
                node = node.predecesor
        path.reverse()
        return path

class Pharmacy():
    def __init__(self, name):
        super.__init__(name)
        self.stocklist = []
```
